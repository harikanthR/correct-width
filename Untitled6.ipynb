```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
gray_img.shape
```

```python
print(gray_img)
print(gray_img.max())
print(gray_img.min())
```

```python
thresh,img_bin = cv2.threshold(gray_img, 100, 255, cv2.THRESH_BINARY)
plt.imshow(img_bin,cmap='gray')
#plt.imshow(img)
```

```python
contours, hierarchy = cv2.findContours(img_bin, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
thickness = 5
img_contours = cv2.drawContours(img.copy(), contours, -1, (0,0,255), thickness)
```
